```python
# Classes in Python

class Point:
    def move(self):
        print("Move")
    def draw(self):
        print("draw")


point_1 = Point()
point_1.x = 10
point_1.draw()
point_1.y= 20
point_1.move()
print(point_1.x, point_1.y)

    
```

```python
# Class with Constructor

class Point:
    def __init__(self, x, y):
            self.x = x
            self.y = y

    def move(self): # the "self" is used to refer to the current instance of the class , which means that it allows you to access the attributes and methods of the class from within
        print("Move")
    def draw(self):
        print("draw")

point = Point(10,20)

print(point.x)
print(point.y)
point.move()

```

```python
# Inheritance

class Mammal:
    def walk(self):
        print("Walk")


class Dog (Mammal):
    def bark(self):
        print("bark")



class Cat(Mammal ): 
    def be_annoying(self):
        print("annoying")


dog = Dog()
dog.bark()
dog.walk()

print()

cat = Cat()
cat.walk()
cat.be_annoying()
```
