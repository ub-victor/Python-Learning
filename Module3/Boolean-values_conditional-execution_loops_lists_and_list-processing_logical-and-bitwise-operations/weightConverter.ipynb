```python
# While lope game, Python use snake_case 
# Variables use a case-sensitive
# Function/Exception use PascalCase eg: MyClass
# Constant use UPPER_CASE_WITH_UNDERSCORE
# Module/Packeges use lowercase

secret_number = 9
guess_count = 0
guess_limit = 3

while guess_count < guess_limit :
    guess = int(input("Guess the secret number: "))
    guess_count+= 1
    if guess != secret_number :
        print("You loose!\n Try again\n")
    elif guess == secret_number :
        print("You won!\n Congratulation")
        break
    else:
        print("Check you input")
else:
    print("You failed all the 3 trials, come back tomorrow")
```

```python
# Car Game
command = ""
started = False
stopped = False
print("""
        start - to start the car
        stop - to stop the car
        quit - to quit
""")

while command != "quit":
    command = input("Enter command > ").lower()
    if command == "start":
        if started:
            print("Car is already Started...")
        else:
            started = True
            print("Car Started...")
        
    elif command == "stop":
        if stopped:
            print("Car is already stopped...")
        else:
            stopped = True 
            print("Car stopped...")
    elif command == "help":
        print("Available commands: start, stop, quit")
    elif command == "quit":
        print("Quitting...")
    else:
        print("Unknown command. Type 'help' for options.")
```

```python
# Emoji Converter :)

message = input("> ")
words = message.split(' ')
# print(words) # it will print an list if the message " , "
emojis = {
    ":)": "😊",
    ":(": "😒"
}

output = ''
for  word in words:
    # print(word) = list
    output+= emojis.get(word, word) + " "
print(output)


```

```python
# Emoji Converter :) with f(x)

def emojis_converter(message) : 
    words = message.split(' ')
    # print(words) # it will print an list if the message " , "
    emojis = {
        ":)": "😊",
        ":(": "😒"
    }

    output = ''
    for  word in words:
        # print(word) = list
        output+= emojis.get(word, word) + " "
    return output

message = input("Enter your msg> ")
emojis_converter(message)

```
