```python
# Let's take a look on pandas series

# Series is line one column i a table

import pandas as pd

data = [10, 20, 30, 40]

s= pd.Series(data) # using this Series will print out a one Column of a table kinda like Key values

print(s)

print("To access data in a Series \n")

print(s[1])

print("Create Series with Labels \n")

s = pd.Series([10, 20, 30], index=["a", "b", "c"])

print(s)
print(s.a) # print(s["a"])
print("Create Series with Labels \n")
```

```python
# Panda DataFrame (This is the most important)
# A DataFrame is a table like Excel

import pandas as pd

data = {
    "Name": ["John", "Anna", "Peter"],
    "Age": [25,22,30],
    "Country": ["USA", "France", "Canada"]
}

df = pd.DataFrame(data)

print(df)

print("Select Columns \n")

print(df["Name"])
print("Select Multiple Columns \n")
print(df[["Name","Country"]])

print("Select Rows \n")

print(df.iloc[0]) # this will print the first row of the table

print("Select Specific Value \n")
print(df.iloc[1,2]) #print("Select Rows \n")

```

```python
# Reading CSV Files(Very Important)
# Most real datasets come from CSV files

import  pandas as pd

df = pd.read_csv("students.csv")

print(df)
print("\nBasic Data Exploration  \n")
print(" Show first rows  \n")

df.head()
print("Show last rows  \n")

df.tail()

print("Show Infomation\n")

df.info()

print("Describe Statistics\n")

df.describe()

print("Filtering Data \n- rows where age > 23\n")
print(df[df["Age"]> 23])

print("Add New Column\n")

df["Salary"] = [222,333,333,55,44,333,33,33,622,555,333]

print(df)
print("Delete a Column\n")

# axis = 1 refers to Columns, and axis= 0 refers to rows
df.drop("Salary", axis=1, inplace=True) # inplace = True means that the change will be made to the original DataFrame without needing to assign it to a new variable
print(df)

print("Sorting Data\n")
print(df.sort_values("Age"))

print("Handling Missing Data\n")
print(df.isnull()) #it respond with boolean

print("Group Data (Intermediate) \n")

# Group the data by age and calculate the mean of each group
print(df.groupby("Age").mean(numeric_only=True))

```
