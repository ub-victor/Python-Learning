```python
import pandas as pd

# Create a simple dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["Paris", "London", "New York"]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
print("My DataFrame:")
print(df)

# Print basic information
print("\nDataFrame Info:")
print(df.info())
```
