```python
# While lope game, Python use snake_case 
# Variables use a case-sensitive
# Function/Exception use PascalCase eg: MyClass
# Constant use UPPER_CASE_WITH_UNDERSCORE
# Module/Packeges use lowercase

secret_number = 9
guess_count = 0
guess_limit = 3

while guess_count < guess_limit :
    guess = int(input("Guess the secret number: "))
    guess_count+= 1
    if guess != secret_number :
        print("You loose!\n Try again\n")
    elif guess == secret_number :
        print("You won!\n Congratulation")
        break
    else:
        print("Check you input")
else:
    print("You failed all the 3 trials, come back tomorrow")
```

```python
# Car Game
command = ""
started = False
stopped = False
print("""
        start - to start the car
        stop - to stop the car
        quit - to quit
""")

while command != "quit":
    command = input("Enter command > ").lower()
    if command == "start":
        if started:
            print("Car is already Started...")
        else:
            started = True
            print("Car Started...")
        
    elif command == "stop":
        if stopped:
            print("Car is already stopped...")
        else:
            stopped = True 
            print("Car stopped...")
    elif command == "help":
        print("Available commands: start, stop, quit")
    elif command == "quit":
        print("Quitting...")
    else:
        print("Unknown command. Type 'help' for options.")
```

```python
# 2 Dimensional Array

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix[0]) # it will out put the first array
print(matrix[0][0]) #it will print out  the first element in the first array
print(matrix) # it will print the whole array
#Modifying the matrice element
matrix[0][1] = 20
print(matrix[0][1])

print("\n")
for row in matrix: 
    for item in row:
        print(item)
    print("\n")


  
```

```python
# List Methods
numbers = [3,4,5,6,6,6,8]
# append us is to add an element at the end of the list
numbers.append(10)
print(numbers)

#  insert gives you the chose to insert at a index choised
numbers.insert(1,12)
print(numbers)

#  remove gives you the chose to remove at a index choised
numbers.remove(6)
print(numbers)


#  clear to delete the whole array
# numbers.clear()
print(numbers)


#  pop  to remove the last index element
numbers.pop()
print(numbers)

print(50 in numbers) # check if a certain number exist in list (boolean as answer)
# print(numbers.index(50)) # check the index of a number but is not mostlu used cuz if the number is not in the list it print an error
print(numbers.count(6)) # to check if the number is repeaded and how many it is repeaded

# To sort a list
numbers.sort()
print(numbers)

# To reverse a list
numbers.reverse()
print(numbers)

# To get copy of  a list
numbers2 = numbers.copy()
numbers.append(10)

print("\n")
print(numbers)
print(numbers2)

```

```python
# Remove the diplicated in a list

numbers = [3, 12, 4, 5, 6, 6, 6, 8, 10]
unique = []

for number in numbers:
    if number not in unique: 
        unique.append(number)
print(unique)
```
