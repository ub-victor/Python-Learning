## 1. What is Matplotlib?

Matplotlib is a Python library used to create graphs and visualizations.

Think of it like:

- Drawing charts (line, bar, pie, etc.)
- Visualizing data clearly
- Used in data science, AI, and research

````python
# Basic Structure 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# x -> Horizotal axis
# y -> Vertical axis
# plot() -> draws the line
# show()

plt.plot(x, y)

# To add Labels & Title

plt.title("Simple line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
````

````python
# Styling the Graph
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

plt.plot(x,y)

plt.title("Simple Line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")


plt.plot(x, y, color = 'red', linestyle = "--", marker="*")

plt.show()
````
