```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
music_data = pd.read_csv('music.csv')

X = music_data.drop(columns=['genre']) # it does not modify the original set, drop() returns a new dataset, music_data is still unchanged
Y = music_data['genre']

model = DecisionTreeClassifier()
model.fit(X,Y)
predictions = model.predict([[21, 1]])
predictions
```
