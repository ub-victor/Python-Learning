```python
# Let's take a look on pandas series

# Series is line one column i a table

import pandas as pd

data = [10, 20, 30, 40]

s= pd.Series(data) # using this Series will print out a one Column of a table kinda like Key values

print(s)

print("To access data in a Series \n")

print(s[1])

print("Create Series with Labels \n")

s = pd.Series([10, 20, 30], index=["a", "b", "c"])

print(s)
print(s.a) # print(s["a"])
print("Create Series with Labels \n")
```

```python
# Panda DataFrame (This is the most important)
# A DataFrame is a table like Excel

import pandas as pd

data = {
    "Name": ["John", "Anna", "Peter"],
    "Age": [25,22,30],
    "Country": ["USA", "France", "Canada"]
}

df = pd.DataFrame(data)

print(df)

print("Select Columns \n")

print(df["Name"])
print("Select Multiple Columns \n")
print(df[["Name","Country"]])

```
