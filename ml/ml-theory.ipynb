# part A - Multiple Choice Question

## 1. What is machine learning inference?
    Machine learning inference is the process of using a trained ML model to make predictions on new data
## 2. What is a trained model?
    A trained model is a machine learning model that has already learned patters from training data
## 3. What does a .pkl file store?
    A .pkl file stores serialized Python objects such as trained ML models
## 4. What is the role of "joblib.load()"
    "Joblib.load()" loads a saved machine learing model form dist into memory
## 5.  What is a feature in ML?
    A feature in ML is an input variable used by the model for prediction
## 6. What is a label in ML?
    A label is the expected output or target value in supervided learning
## 7. What  does predict_proba() return?
    predict_proba returns the probability for each class
