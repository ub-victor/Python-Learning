## 1. What is Seaborn? (Theory)

Seaborn is a Python library used for data visualization.

👉 It is built on top of Matplotlib, but:

- Easier to use
- More beautiful by default
- Works very well with pandas DataFrames
#### Why use Seaborn?
- Less code than Matplotlib
- Built-in themes (pretty graphs)
- Handles statistics + grouping automatically
#### ⚙️ 2. Installation & Setup
pip install seaborn

Then in Python:

import seaborn as sns
import matplotlib.pyplot as plt
🎨 3. Built-in Datasets (SUPER IMPORTANT)

Seaborn comes with datasets (great for exams!)

df = sns.load_dataset("tips")
print(df.head())

Common datasets:

"tips"
"iris"
"titanic"
🎯 4. Core Idea of Seaborn

Seaborn works with variables:

Type	Example
Numeric	age, price
Categorical	gender, day

And maps them to visuals:

x-axis
y-axis
color (hue)
size/style
