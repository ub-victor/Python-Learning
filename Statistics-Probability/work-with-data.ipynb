```python
import pandas as pd

Names  = ["David", "Claudine", "Alice", "Teta"]
Gender = ["Male", "Female", "Female", "Female"]
Height = [1.9, 1.55, 1.66, 1.8]
Weight = [66, 67, 58, 12]

Identifications_2 = {"Names": Names, "Gender": Gender, "Height": Height, "weight": Weight}
df = pd.DataFrame(Identifications_2)
excel = df.to_excel("student_list.xlsx", index=False)
SL = pd.read_excel("student_list.xlsx")


print(df)
print("\n")
SL
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 3))
plt.bar(df["Names"], df["weight"], color= "red")
plt.title("Student distribution by weight", fontsize= 16, color = 'blue')
plt.xlabel("Names", size = 12)
plt.ylabel("Weight", size=12)
plt.xticks(size=11, rotation=0)
plt.yticks(size=11, rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


summary = pd.DataFrame (SL["Gender"].value_counts()).reset_index()
summary.columns = ["Gender", "Counts"]
print(summary)

print("\n")

plt.title("Gender Participation", size=16)
plt.pie(summary["Counts"], labels=summary["Gender"], autopct="%1.1f%%")
plt.show()
```
