## 1. What is Matplotlib?

Matplotlib is a Python library used to create graphs and visualizations.

Think of it like:

- Drawing charts (line, bar, pie, etc.)
- Visualizing data clearly
- Used in data science, AI, and research

````python
# Basic Structure 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# x -> Horizotal axis
# y -> Vertical axis
# plot() -> draws the line
# show()

plt.plot(x, y)

# To add Labels & Title

plt.title("Simple line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
````

````python
# Styling the Graph
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

plt.plot(x,y)

plt.title("Simple Line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")


plt.plot(x, y, color = 'blue', linestyle = "--", marker="*")

plt.show()
````

````python
# Bar Chart
name = ["A", "B", "C"]
values = [5,7,2]

plt.title("Bar Chart")
plt.bar(name, values)
plt.show()

````

````python
# Pie Chart
import matplotlib.pyplot as plt
sized = [40, 30, 20, 10]

labels = ["A", "B", "C", "D"]
# autopct -> to show the percentage of each slice in the pie chart which is represented
# by the "sized" list. The format string '%1.1f%%' specifies that the percentage should 
# be displayed with one decimal place followed by a percent sign.
# The '%%' is used to escape the percent sign in the format string, so it will be displayed as a literal '%' in the output.
# %1.1f -> means to format the number as a floating-point number with one digit after the decimal point.
#  sized -> represents the sizes of each slice in the pie chart, and labels -> represents the labels for each slice.
#  autopct='%1.1f%%' will display the percentage of each slice in the pie chart based on the values in the "sized" list, formatted to one decimal place.
plt.pie(sized, labels=labels, autopct='%1.1f%%')
plt.show()
````

````python
# Scatter Plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 15, 13, 17]

plt.scatter(x, y)

# Used to show relationships between variables
plt.show()
````

````python
# Multiple Lines on One Graph

import matplotlib.pyplot as plt

x = [1, 2, 3]

y1 = [2, 4, 6]
y2 = [1, 3, 5]


plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

# plt.legend() # To show the labels of the lines on the graphq
plt.show()
````
